```python
top_supporters = support.groupby([
    "contributor_lastname", 
    "contributor_firstname"
]).amount.sum().reset_index().sort_values(
    "amount",
    ascending=False
).head(10)
```

```python
chart = top_supporters.amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_lastname)
```

```python
top_supporters["contributor_fullname"] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname

#say if there are spaces in the original headers, you need to rename them with underscores or use square brackets and quotes at all times
```

```python
top_supporters
```

```python
chart = top_supporters.amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_fullname)
```

```python
top_supporters.to_csv("top-supporters.csv", index=False)
```

```python
top_opponents = oppose.groupby([
    "contributor_lastname", 
    "contributor_firstname"
]).amount.sum().reset_index().sort_values(
    "amount",
    ascending=False
).head(10)
```
